```python
# importar as bibliotecas necessárias
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importa o arquivo desejado
nome_do_arquivo = 'HRDataset_v14.csv'
df = pd.read_csv(nome_do_arquivo)
df
```

```python
# 1. Executar estatísticas descritivas básicas no Python
df.describe()
```

```python
# 2. Fazer um histograma com salários
plt.hist(df['Salary'], color='blue', edgecolor='black')  # 'bins' define o número de intervalos no histograma
plt.xlabel('Salários')
plt.ylabel('Frequência')
plt.title('Histograma de Salários')

# Exiba o histograma
plt.show()
```

```python
# criar nova coluna com descrição do sexo
mapeamento = {0: 'Mulher', 1: 'Homem'}
df['Sexo'] = df['GenderID'].map(mapeamento)
```

```python
# 3. fazer um gráfico de violino com salários, discriminados por gênero
sns.violinplot(x='Sexo', y='Salary', data=df, palette='Set3')
plt.xlabel('Gênero')
plt.ylabel('Salário')
plt.title('Distribuição de Salários por Gênero')

# Exiba o gráfico
plt.show()
```

```python
# 3. fazer um gráfico de violino com salários, discriminados por gênero USANDO SEABORN
ax = sns.catplot(data=df, kind='violin', x='Sexo', y='Salary', hue='Sexo', split=True)

ax.set(xlabel='Gênero', ylabel='Salário', title='Distribuição de Salários por Gênero')
```

```python
# Converta a coluna 'dateOfHire' para o formato de data
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Classifique o DataFrame por data de contratação (opcional)
df = df.sort_values(by='DateofHire')

# 4. Traçar a série temporal correspondente à data de contratação
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura conforme necessário
plt.plot(df['DateofHire'], marker='o', linestyle='-')
plt.xlabel('Data de Contratação')
plt.ylabel('Anos')
plt.title('Série Temporal da Data de Contratação')

# Exiba o gráfico

plt.xticks(df['DateofHire'])
plt.grid(True)  # Adicione uma grade opcional
plt.tight_layout()  # Ajuste automático do layout
plt.show()
```
