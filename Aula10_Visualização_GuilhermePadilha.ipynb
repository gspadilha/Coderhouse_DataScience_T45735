```python
# importar as bibliotecas necess√°rias
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importa o arquivo desejado
nome_do_arquivo = 'HRDataset_v14.csv'
df = pd.read_csv(nome_do_arquivo)
df
```

```python
# 1. Executar estat√≠sticas descritivas b√°sicas no Python
df.describe()
```

```python
# 2. Fazer um histograma com sal√°rios
plt.hist(df['Salary'], color='blue', edgecolor='black')  # 'bins' define o n√∫mero de intervalos no histograma
plt.xlabel('Sal√°rios')
plt.ylabel('Frequ√™ncia')
plt.title('Histograma de Sal√°rios')

# Exiba o histograma
plt.show()
```

```python
# criar nova coluna com descri√ß√£o do sexo
mapeamento = {0: 'Mulher', 1: 'Homem'}
df['Sexo'] = df['GenderID'].map(mapeamento)
```

```python
# 3. fazer um gr√°fico de violino com sal√°rios, discriminados por g√™nero
sns.violinplot(x='Sexo', y='Salary', data=df, palette='Set3')
plt.xlabel('G√™nero')
plt.ylabel('Sal√°rio')
plt.title('Distribui√ß√£o de Sal√°rios por G√™nero')

# Exiba o gr√°fico
plt.show()
```

```python
# 3. fazer um gr√°fico de violino com sal√°rios, discriminados por g√™nero USANDO SEABORN
ax = sns.catplot(data=df, kind='violin', x='Sexo', y='Salary', hue='Sexo', split=True)

ax.set(xlabel='G√™nero', ylabel='Sal√°rio', title='Distribui√ß√£o de Sal√°rios por G√™nero')
```

```python
# Converta a coluna 'dateOfHire' para o formato de data
df['DateofHire'] = pd.to_datetime(df['DateofHire'])

# Classifique o DataFrame por data de contrata√ß√£o (opcional)
df = df.sort_values(by='DateofHire')

# 4. Tra√ßar a s√©rie temporal correspondente √† data de contrata√ß√£o
plt.figure(figsize=(10, 6))  # Ajuste o tamanho da figura conforme necess√°rio
plt.plot(df['DateofHire'], marker='o', linestyle='-')
plt.xlabel('Data de Contrata√ß√£o')
plt.ylabel('Anos')
plt.title('S√©rie Temporal da Data de Contrata√ß√£o')

# Exiba o gr√°fico

plt.xticks(df['DateofHire'])
plt.grid(True)  # Adicione uma grade opcional
plt.tight_layout()  # Ajuste autom√°tico do layout
plt.show()
```
