```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando o conjunto de dados
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
column_names = ["ID", "Diagnosis", "Mean Radius", "Mean Texture", "Mean Perimeter", "Mean Area", "Mean Smoothness",
                "Mean Compactness", "Mean Concavity", "Mean Concave Points", "Mean Symmetry", "Mean Fractal Dimension",
                "SE Radius", "SE Texture", "SE Perimeter", "SE Area", "SE Smoothness", "SE Compactness", "SE Concavity",
                "SE Concave Points", "SE Symmetry", "SE Fractal Dimension", "Worst Radius", "Worst Texture", "Worst Perimeter",
                "Worst Area", "Worst Smoothness", "Worst Compactness", "Worst Concavity", "Worst Concave Points", "Worst Symmetry",
                "Worst Fractal Dimension"]
data = pd.read_csv(url, names=column_names)
data
```

```python
# AnÃ¡lise Univariada de Dados: gerar 5 grÃ¡ficos para diferentes caracterÃ­sticas
features_for_univariate = ["Mean Radius", "Mean Texture", "Mean Area", "Mean Compactness", "Mean Symmetry"]

for feature in features_for_univariate:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=data, x=feature, kde=True)
    plt.title(f'DistribuiÃ§Ã£o de {feature}')
    plt.xlabel(feature)
    plt.ylabel("Contagem")
    plt.show()
```

```python
# AnÃ¡lise Bivariada de Dados: gerar 3 grÃ¡ficos - M = malignant, B = benign
# GrÃ¡fico de Barras
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x="Diagnosis", y="Mean Radius")
plt.title("MÃ©dia do Raio por DiagnÃ³stico")
plt.xlabel("DiagnÃ³stico")
plt.ylabel("MÃ©dia do Raio")
plt.show()
```

```python
# GrÃ¡fico de Caixa (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x="Diagnosis", y="Mean Texture")
plt.title("Textura MÃ©dia por DiagnÃ³stico")
plt.xlabel("DiagnÃ³stico")
plt.ylabel("Textura MÃ©dia")
plt.show()
```

```python
# GrÃ¡fico de Barras
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x="Diagnosis", y="Mean Area")
plt.title("Ãrea MÃ©dia por DiagnÃ³stico")
plt.xlabel("DiagnÃ³stico")
plt.ylabel("Ãrea MÃ©dia")
plt.show()
```
