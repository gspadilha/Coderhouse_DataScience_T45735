```python
import pandas as pd
```

```python
# importar o arquivo desejado
nome_do_arquivo = 'us_presidents.csv'
df = pd.read_csv(nome_do_arquivo)
```

```python
df
```

```python
# datas a serem testadas
data_ini = '1850-01-01'
data_fim = '1901-12-31'
```

```python
# converter a data do arquivo de string para datetime
df['start_data'] = pd.to_datetime(df['start'], errors='coerce')
df['end_data'] = pd.to_datetime(df['end'], errors='coerce')
```

```python
# fazer a seleção das data
selecao = df.loc[(df['start_data'] >= data_ini) & (df['start_data'] <= data_fim) | (df['end_data'] >= data_ini) & (df['end_data'] <= data_fim)]
```
