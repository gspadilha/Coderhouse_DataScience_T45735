```python
# Importar as bibliotecas necessárias
import pandas as pd
```

```python
# Ler os arquivos csv
w_data = pd.read_csv("w_data.csv", sep = ';')
p_data = pd.read_csv("p_data.csv", sep = ';')
```

```python
w_data
```

```python
p_data
```

```python
# Fundir os dados a partir da variável ID
merged_data = pd.merge(w_data, p_data, on="ID", how="left")
```

```python
merged_data
```

```python
# Quantos ID têm em comum
common_ids = len(merged_data[merged_data["ID"].notnull()])
print(f"O número de IDs em comum é {common_ids}")
```

```python
# Nomear quais não têm em comum
missing_ids = merged_data[merged_data["Age"].isnull()]["Age"]
print(f"O número de IDs que NÃO são comuns é {len(missing_ids)}:\n{missing_ids}")
```

```python
merged_data['Target'].unique()
```

```python
# Corrigir os target a apenas dois valores distintos
merged_data["Target"] = merged_data["Target"].str.strip().str.upper()
merged_data["Target"] = merged_data["Target"].replace(["<=50K.", ">50K."], ["<=50K", ">50K"])
merged_data
```

```python
# Calcular a proporção entre pessoas com salários >50k em relação ao total com pessoas que difiram em sexo, raça e nível educacional
def calc_proportion(by):
    counts = merged_data.groupby([by, "Target"]).size().unstack()
    ratio = counts[">50K"] / counts.sum(axis=1)
    print(f"A proporção de pessoas com salários >50K por {ratio}")
```

```python
# Por Sexo
calc_proportion("Sex")
```

```python
# Por Raça
calc_proportion("Race")
```

```python
# Por Educação
calc_proportion("Education")
```
