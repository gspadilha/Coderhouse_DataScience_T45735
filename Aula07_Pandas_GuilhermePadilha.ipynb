```python
import pandas as pd
```

```python
# importa o arquivo desejado
nome_do_arquivo = 'Summer-Olympic-medals-1976-to-2008.csv'
df = pd.read_csv(nome_do_arquivo, sep=',', encoding='latin-1')
df
```

```python
# copiar o DF para não perder o original
new_df = df.copy()
# transformar o ano em inteiro
new_df['Year'] = new_df['Year'].fillna(0).astype(int)
# definir a coluna que será o índice
new_df.set_index('Year', inplace=True)

#1. construir um objeto de Serie cronológica com o índice igual ao ano dos jogos olímpicos
new_df
```

```python
# separar os nomes em duas colunas distintas
sobrenome_nome = new_df['Athlete'].str.split(',', n=1, expand=True)
# renomear as novas colunas
sobrenome_nome.columns = ['Sobrenome', 'Nome']

#@TODO: como inverter as colunas antes de concatenar

# concatenar as novas colunas com o DF
new_df = pd.concat([new_df, sobrenome_nome], axis=1)

#2. Separar os nomes próprios de sobrenomes em duas colunas distintas usando vírgula como separador.
new_df
```

```python
# agrupar os dados por ano e país, somar a quantidade de medalhas (@TODO: como contar por tipo de medalhar?)
mais_medalhas_por_ano = new_df.groupby(['Year', 'Country'])['Medal'].size().reset_index()
# retorna um booleano indicando se o valor é igual ao MAX do ano
idx = mais_medalhas_por_ano.groupby('Year')['Medal'].transform(max) == mais_medalhas_por_ano['Medal']
# mostrar os vencedores
paises_vencedores_por_ano = mais_medalhas_por_ano[idx]

#3. Obter medidas resumidas do conjunto de dados. Qual país ganhou mais medalhas (por ano)?
paises_vencedores_por_ano
```

```python
pais_vencedor = mais_medalhas_por_ano.loc[mais_medalhas_por_ano['Medal'].idxmax()]

#3. Obter medidas resumidas do conjunto de dados. Qual país ganhou mais medalhas (em todos os anos)?
pais_vencedor
```

```python
def filtrar_e_contar(df, sexo = 'A'): # A -> Todos, M -> Homens, W -> Mulheres
    df_filtrado = df

    # caso não seja todos, filtrar pelo sexo indicado
    if sexo != 'A':
        df_filtrado = df[df['Gender'].astype(str).str[0] == sexo]

    # agrupar os dados por ano, somar a quantidade de medalhas (@TODO: passar um array de agrupamento)
    medalhas_por_ano = df_filtrado.groupby(['Year'])['Medal'].size().reset_index()
    # retorna um boolean indicando se o valor é igual ao MAX do ano
    index = medalhas_por_ano.groupby('Year')['Medal'].transform(max) == medalhas_por_ano['Medal']
    # retornar os valores
    return medalhas_por_ano[index]
```

```python
#4. Construir uma tabela mostrando quantas medalhas foram ganhas pelos homens no total em cada ano em que o evento foi realizado.
medalhas_homens = filtrar_e_contar(new_df, 'M')
medalhas_homens
```

```python
medalhas_mulheres = filtrar_e_contar(new_df, 'W')
medalhas_mulheres
```
